```python
from pathlib import Path

def create_json(data):
    return "{}"

for i in range(10):
    for j in range(5):
        p = Path("sample_data", f"folder-{i}", f"file-{j}.json")
        if not p.parent.exists():
            p.parent.mkdir(parents=True, exist_ok=True)
            p.write_text(create_json(i * j))
```
